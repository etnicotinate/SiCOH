```python
from collections import defaultdict

# Initialize dictionaries to store bond information
si_o_bonds = defaultdict(int)

# Read the bond data file
with open('bonds.test', 'r') as file:
    lines = file.readlines()
    for line in lines:
        # Skip the header line
        if line.startswith('#'):
            continue
        # Split the line into columns
        columns = line.split()
        bond_id, atom1_type, atom2_type, atom1_id, atom2_id, bond_order, atom1_notation, atom2_notation = columns
        if atom1_notation == 'Si' and atom2_notation == 'O':
            si_o_bonds[atom1_id] += 1
        elif atom1_notation == 'O' and atom2_notation == 'Si':
            si_o_bonds[atom2_id] += 1

# Initialize counters for different numbers of O atoms bonded to Si
Q = 0  # Si with 1 O
T = 0  # Si with 2 O
D = 0  # Si with 3 O
M = 0  # Si with 4 O

# Count the Si atoms with 1, 2, 3, and 4 O atoms
for si_id, o_count in si_o_bonds.items():
    if o_count == 1:
        Q += 1
    elif o_count == 2:
        T += 1
    elif o_count == 3:
        D += 1
    elif o_count == 4:
        M += 1

# Output the counts
print(f"Si with 1 O (Q): {Q}")
print(f"Si with 2 O (T): {T}")
print(f"Si with 3 O (D): {D}")
print(f"Si with 4 O (M): {M}")
```

```python
from collections import defaultdict

# Initialize dictionaries to store bond information
si_o_bonds = defaultdict(int)

# Read the bond data file
with open('bonds.test', 'r') as file:
    lines = file.readlines()
    for line in lines:
        # Skip the header line
        if line.startswith('#'):
            continue
        # Split the line into columns
        columns = line.split()
        bond_id, atom1_type, atom2_type, atom1_id, atom2_id, bond_order, atom1_notation, atom2_notation = columns
        if atom1_notation == 'Si' and atom2_notation == 'O':
            si_o_bonds[atom1_id] += 1
        elif atom1_notation == 'O' and atom2_notation == 'Si':
            si_o_bonds[atom2_id] += 1

from collections import defaultdict

# Initialize dictionaries to store bond information
si_o_bonds = defaultdict(int)

# Read the bond data file
with open('bonds.test', 'r') as file:
    lines = file.readlines()
    for line in lines:
        # Skip the header line
        if line.startswith('#'):
            continue
        # Split the line into columns
        columns = line.split()
        bond_id, atom1_type, atom2_type, atom1_id, atom2_id, bond_order, atom1_notation, atom2_notation = columns
        if atom1_notation == 'Si' and atom2_notation == 'O':
            si_o_bonds[atom1_id] += 1
        elif atom1_notation == 'O' and atom2_notation == 'Si':
            si_o_bonds[atom2_id] += 1

# Initialize QTD dictionary
cur_QTD = {'Q': [], 'T': [], 'D': [], 'M': []}

# Categorize Si atoms into Q, T, D, and M groups
for si_id, o_count in si_o_bonds.items():
    if o_count == 1:
        cur_QTD['M'].append(si_id)
    elif o_count == 2:
        cur_QTD['D'].append(si_id)
    elif o_count == 3:
        cur_QTD['T'].append(si_id)
    elif o_count == 4:
        cur_QTD['Q'].append(si_id)

# Count the number of Si atoms in each group
QTD_counts = {c: len(ids) for c, ids in cur_QTD.items()}

# Output the counts
print(f"Si with 1 O (M): {QTD_counts['M']}")
print(f"Si with 2 O (D): {QTD_counts['D']}")
print(f"Si with 3 O (T): {QTD_counts['T']}")
print(f"Si with 4 O (Q): {QTD_counts['Q']}")

# Output the QTD dictionary for verification
print(f"QTD Dictionary: {cur_QTD}")
```
