```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for demonstration, replace this with your actual rdf_data
rdf_data = [np.random.rand(5, 10) for _ in range(2)]  # Random data for example

# Extract the 1st frame (timestep 0)
data = rdf_data[0]

labels = ['C-C', 'C-Si', 'Si-O']
colors = ['blue', 'green', 'red']
r = data[:, 0]  # distance, r

ax = plt.gca()

# Plot initial timestep curves
lines_0 = []
for i, (label, color) in enumerate(zip(labels, colors)):
    g_r = data[:, 2*i+1]  # g(r)
    line, = plt.plot(r, g_r, label=f'{label}', color=color, linestyle='-')
    lines_0.append(line)

# Create a legend for the initial timestep
legend1 = plt.legend(handles=lines_0, loc='upper left')
plt.gca().add_artist(legend1)

# Plot the middle frame (timestep 1000)
data = rdf_data[1]
lines_1000 = []
for i, (label, color) in enumerate(zip(labels, colors)):
    g_r = data[:, 2*i+1]  # g(r)
    line, = plt.plot(r, g_r, label=f'{label}', color=color, linestyle='--')
    lines_1000.append(line)

handles, labels = ax.get_legend_handles_labels()
handles_middle = handles[len(labels)//2:]
labels_middle = labels[len(labels)//2:]
legend2=plt.legend(handles=handles_middle, labels=labels_middle, loc='upper right', title='Timestep 1000')


plt.xlabel('r')
plt.ylabel('g(r)')
plt.title('RDF Comparison at Timestep 0 and Timestep 1000')

```
